# 가사 크롤링 해오기!

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import time
```

```python
html = urlopen('https://music.bugs.co.kr/genre/kpop/rnh/total?tabtype=2')
soup_tmp = BeautifulSoup(html, 'html.parser')
soup_tmp
```

```python
soup_tmp.find_all('a', class_='trackInfo')
```

```python
soup_tmp.find_all('a', class_='trackInfo')[0]
```

```python
soup_tmp.find_all('a', class_='trackInfo')[0]['href']
```

```python
total_lyrics=''
for i in range(0,50):
    lyrics_url = soup_tmp.find_all('a', class_='trackInfo')[i]['href']
    url_html = urlopen(lyrics_url)
    html_parsed = BeautifulSoup(url_html, 'html.parser')
    lyrics = html_parsed.find_all(class_='lyricsContainer')[0].get_text()
    total_lyrics += lyrics
    time.sleep(2)

total_lyrics
```

```python
import re
lyrics_ = re.sub('[\n\r\t]', ' ', total_lyrics)
```

```python
lyrics_
```

```python
lyrics__ = re.sub('가사 오류 제보',"",lyrics_)
```

```python
lyrics__ = re.sub('님이 등록해 주신 가사입니다.',"",lyrics__)
```

```python
lyrics__ = re.sub(r"\\'","'",lyrics__)
```

```python
lyrics__
```

# 가사 분석하기!

```python
from konlpy.tag import Twitter
```

```python
twitter = Twitter()
by_n = twitter.nouns(lyrics__)
```

```python
by_n
```

```python
twitter.
```

```python
from collections import Counter
```

```python
cnt=Counter(by_n)
cnt
```

```python
cnt_dict={word : cnt[word] for word in cnt}
```

```python
cnt_df = pd.DataFrame.from_dict(cnt_dict, orient='index')
```

```python
cnt_df
```

```python
df_sorted = cnt_df.sort_values(0,ascending=False)
```

```python
top50words = df_sorted.head(50)
top50words
```
