```python
import cv2
import sys
```

```python
image_file = 'c:/users/ximzz/desktop/face/IMG_1380.jpg'
```

```python
cascade_file = 'C:/Users/ximzz/Anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_alt.xml'
```

```python
image = cv2.imread(image_file)
image_gs = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cascade = cv2.CascadeClassifier(cascade_file)
```

```python
face_list = cascade.detectMultiScale(image_gs,
                                     scaleFactor=1.1, 
                                     minNeighbors=1, 
                                     minSize=(150,150))
```

```python
if len(face_list) >0:
    print(face_list)
    color= (0,0,255)
    for face in face_list:
        x,y,w,h = face
        print(x,y,w,h)
        cv2.rectangle(image, (x,y-200), (x+w, y+h+int(h*0.28)), color, thickness=8)
    
    cv2.imwrite('facedetect-output.PNG', image)
    
else :
    print('no face')
```

# upperbody_detection

```python
cascade_file = 'C:/Users/ximzz/Anaconda3/Lib/site-packages/cv2/data/haarcascade_upperbody.xml'
```

```python
image = cv2.imread(image_file)
image_gs = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cascade = cv2.CascadeClassifier(cascade_file)
```

```python
body_list = cascade.detectMultiScale(image_gs,
                                     scaleFactor=1.1, 
                                     minNeighbors=1, 
                                     minSize=(150,150))
```

```python
if len(body_list) >0:
    print(body_list)
    color= (0,0,255)
    for body in body_list:
        x,y,w,h = body
        print(x,y,w,h)
        cv2.rectangle(image, (x,y), (x+w, y+h), color, thickness=8)
    
    cv2.imwrite('bodydetect-output.PNG', image)
    
else :
    print('no body')
```
